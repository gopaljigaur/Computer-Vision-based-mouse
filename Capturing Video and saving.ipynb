```python

import cv2 as cv
import numpy as np
```

```python
#from camera
cam = cv.VideoCapture(0)

while(True):
    #capture frame-by-frame
    ret, frame = cam.read() #bool ret=if frame found, frame is image
    print(ret)
    #display resulting image 'frame'
    cv.imshow('Frame',frame)
    
    key = cv.waitKey(10)
    if key==27:
        break
#finally
cam.release()
cv.destroyAllWindows()
    
```

```python
#from video

vid = cv.VideoCapture('vtest.avi')

while(vid.isOpened()): #if video is opened in the window
    ret, frame = vid.read()
    if ret:
        cv.imshow('Video from file',frame)
    key = cv.waitKey(80)
    if key == 27 or ret == False:
        break
        
vid.release()
cv.destroyAllWindows()
```

```python
#saving video file
cam1 = cv.VideoCapture(0)
#define the codec and create VideoWriter object
name = 'camvid.avi'
#defining the codec
fourcc = cv.VideoWriter_fourcc(*'XVID')
#videowriter object: name, codec, framerate, window size
output = cv.VideoWriter(name, fourcc, 30.0, (640,480))

while(cam1.isOpened()):
    ret, frame = cam1.read()
    if ret == True:
        cv.imshow('Video', frame)
        output.write(frame)
        
        if cv.waitKey(1) == ord('q'):
            print('Video file saved as',name)
            break
    else:
        print('Error in video capture')
        break
        
cam1.release()
output.release()
cv.destroyAllWindows()
        
```
